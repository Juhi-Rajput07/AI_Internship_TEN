## What is EDA?
Exploratory Data Analysis (EDA) is an important step in data analysis where we explore, summarize, and visualize data to understand its structure, detect patterns, identify anomalies, test assumptions, and check relationships between variables before applying any machine learning or statistical models.

## Why EDA?
Exploratory Data Analysis (EDA) is crucial because raw data is messy, unpredictable, and full of hidden traps that can ruin your analysis or machine learning models. Skipping EDA is like building a house without checking the foundation—it saves time upfront but risks catastrophic failure later.

## Why do EDA??
* Model building
* Analysis and reporting
* Validate assumptions
* Handling missing values
* feature engineering
* detecting outliers

## Types of EDA
EDA is divided into four types based on how many variables you look at and whether you use numbers or charts:

1. Univariate Non-Graphical
* What it does: Looks at one variable using numbers.
* Examples: Calculating the average age, median income, or standard deviation.

2. Univariate Graphical
* What it does: Looks at one variable using charts.
* Examples: Using a Histogram to see data shape or a Box Plot to find outliers.

3. Multivariate Non-Graphical
* What it does: Looks at multiple variables using numbers.
* Examples: A Correlation Matrix showing how strongly two variables relate to each other.

4. Multivariate Graphical
* What it does: Looks at multiple variables using charts.
* Examples: A Scatter Plot to see trends or a Heatmap to spot visual correlations.
