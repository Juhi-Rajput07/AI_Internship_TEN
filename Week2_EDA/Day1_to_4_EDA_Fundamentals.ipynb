# Day 1 - EDA  Fundamentals  (Tutorial)

## What is EDA?
Exploratory Data Analysis (EDA) is an important step in data analysis where we explore, summarize, and visualize data to understand its structure, detect patterns, identify anomalies, test assumptions, and check relationships between variables before applying any machine learning or statistical models.

## Why EDA?
Exploratory Data Analysis (EDA) is crucial because raw data is messy, unpredictable, and full of hidden traps that can ruin your analysis or machine learning models. Skipping EDA is like building a house without checking the foundation—it saves time upfront but risks catastrophic failure later.

## Why do EDA??
* Model building
* Analysis and reporting
* Validate assumptions
* Handling missing values
* feature engineering
* detecting outliers

## Types of EDA
EDA is divided into four types based on how many variables you look at and whether you use numbers or charts:

1. Univariate Non-Graphical
* What it does: Looks at one variable using numbers.
* Examples: Calculating the average age, median income, or standard deviation.

2. Univariate Graphical
* What it does: Looks at one variable using charts.
* Examples: Using a Histogram to see data shape or a Box Plot to find outliers.

3. Multivariate Non-Graphical
* What it does: Looks at multiple variables using numbers.
* Examples: A Correlation Matrix showing how strongly two variables relate to each other.

4. Multivariate Graphical
* What it does: Looks at multiple variables using charts.
* Examples: A Scatter Plot to see trends or a Heatmap to spot visual correlations.

## Step 5: Ask a Question about the data
* Try to answer a question you have about the data using a plot or statistic.
  
What are the locations with the fastest roller coasters (minimum of 10)?

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```

#  Day 3 --- Understanding Relationship in Data

## BOX PLOT 
* Median: The middle value; it shows the center of your data.
* Spread: The range and box width; it shows data variability.
* Outliers: Individual dots; they show extreme, unusual data points.

```python
ax = sns.boxplot(x = df['Speed_mph'],
                data=df)
ax.set_title('Coaster Speed')         #Univariate Boxplot 
plt.show()
```

```python
# OBSERVATIONS
# The median coaster speed is approximately 50 mph.
# Most coaster speeds are concentrated between approximately 37 mph and 57 mph.
# Several high-speed roller coasters appear as outliers above the normal speed range.(90-150mph)
# A few unusually slow coasters are also present.---> at the left side (5-7mph)
```

## SCATTER PLOT
Relations in scatter plot 
* Positive: Dots go up and right. Both variables increase together.
* Negative: Dots go down and right. One increases, the other decreases.
* No Relation: Dots are a random cloud. Changing X does not affect Y.

```python
ax = sns.scatterplot(x='Speed_mph',
                    y='Height_Ft',
                    data=df)
ax.set_title('Coaster Speed vs Height')
plt.show()
```

## CORRELATION MATRIX

```python
df.select_dtypes(include='number').columns
```

```python
df_corr = df.select_dtypes(include='number').corr()
df_corr
```

### Correlation Analysis
* Speed_mph and Height_Ft have the strongest positive correlation (0.823), indicating that taller roller coasters generally tend to have higher speeds.
* Latitude and Longitude have the strongest negative correlation (-0.298), but the relationship is weak.
* Most other numerical features have weak correlations, suggesting they are relatively independent of each other.

## HEATMAP

```python
sns.heatmap(df_corr, annot= True)
```

### Heatmap Observations

1. Speed_mph and Height_Ft have the strongest positive correlation (0.823), indicating that taller roller coasters generally tend to be faster.

2. Latitude and Longitude have the strongest negative correlation (-0.298), although the relationship is weak.

3. Gforce has a moderate positive correlation with Speed_mph (0.50), suggesting that higher-speed coasters often experience greater g-forces.

4. Latitude and Inversions have almost no correlation (-0.0005), indicating little to no linear relationship.

5. The heatmap makes it easier to identify relationships between numerical variables compared to reading the correlation matrix alone.
