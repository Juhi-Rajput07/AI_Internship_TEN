## TASK 3  -- Data Cleaning

```python
#tit.isna()       #check missing values-(NaN, None, NaT)

tit.isna().sum()  #This shrinks those 891 rows of confusing booleans into a single, clean list of numbers.
```

```python
tit[tit['Age'].isna()]     #this can give the exact no of rows that are broken
```

```python
# tit.dropna()    # remove missing values -- this is a bad idea this can loose ur entire datset

tit['Cabin'] = tit['Cabin'].fillna('unknown')   #hnadling cabin column

median_age = tit['Age'].median()              # handle age column
tit['Age'] = tit['Age'].fillna(median_age) 

tit.dropna(subset=['Embarked'], inplace =True)   #hnadling embarked column

tit.isna().sum()
```

```python
# CHECKING DUPLICATES

# tit.duplicated()             # --detecting
tit.duplicated().sum()     #-- how many duplicates
```

### Observation
* Age has 177 missing values.
* Cabin has 687 missing values.
* Embarked has 2 missing values.

## TASK 4  -- Data Analysis

```python
# HOW MANY PASSENGERS SURVIVED?
tit['Survived'].sum()
```

```python
# HOW MANY PASSENGERS DID NOT SURVIVE?
(tit['Survived'] == 0).sum()
```

```python
# How many male and female passengers were there?
# tit['Sex'].count()
tit['Sex'].value_counts()  # value_counts this is good for categorical data 
```

```python
# How many passengers belonged to:
# First Class
# Second Class
# Third Class

tit['Pclass'].value_counts()
```

```python
#Average age of passengers.
tit['Age'].mean()
```

```python
# Average fare paid.
tit['Fare'].mean()
```

```python
# Which gender had a higher survival count?

 ##---- To analyze two columns together, use double square brackets [[ 'Col1', 'Col2' ]] to select them as a new DataFrame

tit.groupby('Sex')['Survived'].sum()     #in ()bracket we add that column jis category ko lekr pucha gyauska col, nd []bracket mei woh jo pucha gya use related column 
# ans is female
```

```python
# Which passenger class had the highest survival count?
tit.groupby('Pclass')['Survived'].sum()
# ans is 1st class
```

```python
# Passengers older than 60.
# How many?

(tit['Age']>60).sum()
```

```python
# Passengers younger than 18.
# How many?

(tit['Age']<18).sum()
```

### Observation
* Only 340 passengers survived out of 891.
* Number of male were more than female .
* Majority of the passenger are from third class.
* Average age of passenger was approximately 29.
* Female passengers survived more often than male.
* The passengers belong to 1st class had highest survival count.
* There were 113 children and 21 senior citizen.
  

### Research
#### Why is Pandas important for Data Analysis?
Pandas is important because it turns slow, messy data tasks into fast, automated code. It saves time, prevents mistakes, and gives you total control over your data.

#### Why companies use Pandas.
* Saves Money: Automates boring, repetitive manual reporting.
* Builds Trust: Code prevents human copy-paste errors.
* Saves Knowledge: Scripts show exactly how metrics were calculated

#### Where Data Analysts use it.
* Cleaning Data: Fixing messy text, missing numbers, and duplicates.
* Exploring Trends: Finding quick averages, patterns, and anomalies.
* Merging Sources: Combining different files into one final report.

#### Advantages over Excel.
* Never Crashes: Handles millions of rows without freezing.
* Easy Auditing: Shows step-by-step logic instead of hidden cell formulas.
* One-Click Run: Re-runs weekly reports instantly with one button.

#### Why it is useful for large datasets.
* Blazing Fast: Uses optimized speed to calculate columns simultaneously.
* Lightweight Memory: Compresses huge files to fit on standard laptops.
* No Row Limits: Breaks past Excel's strict 1-million-row ceiling.
