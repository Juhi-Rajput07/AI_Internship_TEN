## TASK 3  -- Data Cleaning

```python
#tit.isna()       #check missing values-(NaN, None, NaT)

tit.isna().sum()  #This shrinks those 891 rows of confusing booleans into a single, clean list of numbers.
```

```python
tit[tit['Age'].isna()]     #this can give the exact no of rows that are broken
```

```python
# tit.dropna()    # remove missing values -- this is a bad idea this can loose ur entire datset

tit['Cabin'] = tit['Cabin'].fillna('unknown')   #hnadling cabin column

median_age = tit['Age'].median()              # handle age column
tit['Age'] = tit['Age'].fillna(median_age) 

tit.dropna(subset=['Embarked'], inplace =True)   #hnadling embarked column

tit.isna().sum()
```
