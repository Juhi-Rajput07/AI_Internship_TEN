## What is Pandas?
Pandas is a software library written for the Python programming language for data manipulation and analysis.In particular, it offers data structures and operations for manipulating numerical tables and time series.
## Why Pandas?
Pandas is popular because it works like a super-powered Excel inside Python.
Think of it as Excel, but you type instructions instead of clicking buttons.
Here is why people use it:
* It opens files: It loads data from Excel or text files instantly.
* It cleans data:It fixes typos and fills in missing numbers easily.
* It sorts info: You can filter and organize big lists in seconds.
* It does math: It calculates totals and averages for you automatically.
* It handles huge files: It reads massive files that would crash normal Excel.

## Pandas main Data-structures
* Dataframes --A DataFrame has both columns and rows,is the whole table.
* Series --A Series is just one single column of data, but it still has rows inside that column

## Importing data using read_csv()

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('matches.csv')
```

```python
data
```

```python
# The head() tool shows you the very beginning of your data.
# By default, it grabs the first 5 rows of your table so you can quickly look at it without loading millions of lines.

data.head()   
```

```python
data.head(3)
```

```python
# The tail() tool shows you the very end of your data.It is the exact opposite of head().
# By default, it grabs the last 5 rows of your table so you can check how your data ends.

data.tail()
```

```python
data.tail(1)
```

```python
# It tells you the exact size of your table by giving you two numbers: (Rows, Columns).

data.shape
```

```python
# The info() tool gives you a complete overview of your dataset.
# Unlike shape, you do need the parentheses here (df.info()) because it performs an action to scan your table.

data.info()
```

```python
#The describe() tool gives you a quick mathematical summary of all the numeric columns in your table.

data.describe()
```

```python
data['winner']    # single column
# type(data['winner') ---> type will series
```

```python
data[['team1', 'team2', 'winner']]   # multiple column
# ---> type will dataframe
```

```python
data['winner'].shape  
```

```python
data[['team1', 'team2', 'winner']].shape
```

```python
#  .iloc does not use parentheses (). Instead, it uses square brackets [] because you are indexing (pointing to) specific locations. You type it like df.iloc[...].
# The "i" stands for integer, meaning it finds data using exact row and column numbers (starting from 0), just like coordinates on a map.
data.iloc[0]   #-->series
```

```python
data.iloc[1:3]  #-->dataframe
```

```python
data.iloc[1:11:2]   #---> slices your table to grab every second row starting from index 1 up to (but not including) index 11.It
```

```python
# facy indexing
data.iloc[[1,5,6]] 
```

```python
data.iloc[:,[4,5,10]]
```
