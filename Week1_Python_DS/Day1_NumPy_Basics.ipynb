## What is NumPy?
NumPy(NUmerical Python) is a python library used for numerical computing.It provides a powerful multidimensional array(ndarray) and many mathematical functions.
## Why NumPy?
- Faster than Python lists
- Less memory usage
- Supports vectorized operations
- Used in Data Science and Machine Learning

```python
x= range(100000000)
y=range(100000000, 200000000)

start_time = time.time()

c = [x+y for x,y in zip(x,y)]

print(time.time()-start_time)
```

```python
a = np.arange(100000000)
b = np.arange(100000000, 200000000)

start_time = time.time()
c = a + b

print(time.time()-start_time)
```

# Indexing, SLicing and Iteration

```python
arr12 = np.arange(24).reshape(6,4) #(rows,col)--factors of 24 in matrix
arr12
```

```python
arr12[2] # this will give a row
```

```python
arr12[:,1:4] # thi will give coloumn  
```

```python
arr12[2:4,1:3]
```

```python
 for i in arr12: # this will print whole item in rows
     print(i) 
```

```python
# iteration
for i in np.nditer(arr12):     # this will print whole item in columns
    print(i)
```

## (Student Mark Analysis)

```python
import numpy as np
```

```python
marks = np.array([48, 37, 95, 89, 67, 98])
print("Original marks = ", marks)
```

```python
print("Total number of students = ", marks.size)
```

```python
marks[0] # marks of first student
```

```python
print("MArks of last student = ", marks[5]) #marks of last one
```

```python
marks[:4] # marks of first four students 
```

```python
marks[-3:] # last three students
```

```python
marks + 5 # Increase every students marks by 5
```

```python
marks * 2  #multiply by 2 to evry marks
```

## Observation

NumPy performs mathematical operations on the entire array at once without using loops, making the code simpler and faster.
